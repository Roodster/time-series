```python
#**********************************************************************
#Slide-32: Logarithmic/Square root/Box-Cox transform for Fixing Non-constant Variance
#**********************************************************************
from pandas import read_csv
from numpy import log
from numpy import sqrt
from scipy.stats import boxcox
import matplotlib.pyplot as plt
```

```python
path = 'datasets/airline-passengers.csv'
```

```python
series = read_csv(path, header=0, parse_dates=[0], index_col=0, squeeze=True)
series_log = log(series)
series_sqr = sqrt(series)
series_boxcox, lam= boxcox(series)
print('Lambda: %f' % lam)
plt.plot(series_log)
plt.plot(series_sqr)
plt.plot(series_boxcox)
plt.show()
```
